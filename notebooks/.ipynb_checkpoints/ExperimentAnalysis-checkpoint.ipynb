```python
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
reductionMethods = ['MaxPool','AvgPool','Strided','Dilated']
filterNums = ['04','08','16','32','64','128']
test_results = {}
learning_curves = {}
training_curves = {}
validation_curves = {}
best_val = {}
for num in filterNums:
    for method in reductionMethods:
        name = 'f'+num+method
        test_results[name] = np.genfromtxt('data/filterData/'+ name +'/result_outputs/test_summary.csv', 
                                       skip_header= 1,delimiter=',')
        temp = test_results[name];
        test_results[name] = temp[0]
        learning_curves[name] = np.genfromtxt('data/filterData/'+name+'/result_outputs/summary.csv', 
                                       skip_header= 1,delimiter=',')
        temp = learning_curves[name];
        training_curves[name] = temp[:,0]
        validation_curves[name] = temp[:,2]
        best_val[name] = np.max(validation_curves['f'+num+method])
```

```python
for num in filterNums:
    for method in reductionMethods:
        name = 'f'+num+method
        print(name,best_val[name])
    print()
```

```python
for num in filterNums:
    for method in reductionMethods:
        print('f'+num+method,test_results['f'+num+method])
    print()
```

```python
method = 'MaxPool'
for num in filterNums:
    y = training_curves['f'+num+method]
    x = np.arange(y.shape[0])
    plt.plot(x,y,label = 'f'+num+method)
plt.legend()
```

```python
num = '32'
for method in reductionMethods:
    y = training_curves['f'+num+method]
    x = np.arange(y.shape[0])
    plt.plot(x,y,label = 'f'+num+method+' Train')
plt.legend()
```

```python
num = '32'
for method in reductionMethods:
    y = validation_curves['f'+num+method][:]
    x = np.arange(y.shape[0]) 
    plt.plot(x,y,label = 'f'+num+method+' Val')
plt.legend()
```
